```python
directory = "../Results/"
csv_files = []
for root, dirs, files in os.walk(directory):
    for file in files:
        if file.endswith("energy.csv"):
            csv_files.append(os.path.join(root, file))

# avg client power	avg server power	time	computer	energy	site
                
dataframes = []
for csv_file in csv_files:
    try:
        df = pd.read_csv(csv_file)
        dataframes.append(df)
    except Exception as e:
        print(f"Error importing {csv_file}: {e}")

df = pd.concat(dataframes, ignore_index=True)
display(df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df_groupby_computer = df.groupby('computer')

for computer, df_computer in df_groupby_computer:
    df_groupby_plugin = df.groupby('plugin')
    for plugin, df_plugin in df_groupby_plugin:
        df_groupby_theme = df_plugin.groupby('theme')
        for theme, df_theme in df_groupby_theme:
            x = df_theme['time']
            y = df_theme['energy']
            res = stats.linregress(x, y)
            print(res)
            print(f"R-squared: {res.rvalue**2:.6f}")
            plt.plot(x, y, 'o', label='original data')
            plt.plot(x, res.intercept + res.slope*x, 'r', label='fitted line')
            plt.legend()
            plt.show()

            print(computer,theme)
            sns.lmplot(data=df_theme, x='time',y='energy', hue='plugin', palette='colorblind')
            plt.show()
```

```python
# create medians overview
df_groupby_computer = df.groupby('computer') # client or server
time_list = []
energy_list = []
power_list = []

for computer, df_computer in df_groupby_computer:
    df_groupby_plugin = df.groupby('plugin') # client or server
    for plugin, df_plugin in df_groupby_plugin:
        df_groupby_theme = df_plugin.groupby('theme')
        for theme, df_theme in df_groupby_theme:
            time_median = df_theme['time'].median()
            energy_median = df_theme['energy'].median()
            if computer == "client energy":
                power_median = df_theme['avg client power'].median()
                comp = 'client'
            else:
                power_median = df_theme['avg server power'].median()
                comp = 'server'
            print(str(comp)+','+str(plugin)+
                  ','+str(theme)+','+str(time_median)+','+str(energy_median)+','+
                  str(power_median))
            
```
